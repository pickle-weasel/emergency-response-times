```python
import os
import csv
import pandas as pd
```

```python
meta_file = 'd11_text_meta_2018_10_31.txt'

meta_data = []

with open (meta_file, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    next(csvreader)
    for row in csvreader:
        meta_data.append(row[0].split('\t'))
        
meta_dict = {
            'ID': [],
            'Latitude': [],
            'Longitude': [],
            'Name': []
            }
        
for data in meta_data:
    meta_dict['ID'].append(data[0])
    meta_dict['Latitude'].append(data[8])
    meta_dict['Longitude'].append(data[9])
    meta_dict['Name'].append(data[13])
```

```python
metadata_df = pd.DataFrame(meta_dict)
metadata_df.head()
```

```python
sensor_file = 'd11_text_station_5min_2018_10_31.txt'

traffic_data = {
                'Timestamp': [],
                'ID': [],
                'District': [],
                'Freeway': [],
                'Direction': [],
                'Lane Type': [],
                'Avg Speed': []
                }

with open (sensor_file, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    for row in csvreader:
        traffic_data['Timestamp'].append(row[0])
        traffic_data['ID'].append(row[1])
        traffic_data['District'].append(row[2])
        traffic_data['Freeway'].append(row[3])
        traffic_data['Direction'].append(row[4])
        traffic_data['Lane Type'].append(row[5])
        traffic_data['Avg Speed'].append(row[11])       
```

```python
traffic_df = pd.DataFrame(traffic_data)
traffic_df.head()
```

```python
d11_complete = pd.merge(metadata_df, traffic_df, how='left', on=['ID'])
```

```python
d11_complete = d11_complete.loc[d11_complete['Lane Type'] == 'ML']
d11_complete.reset_index(drop=True, inplace=True)
d11_complete.head()
```

```python
d11_complete.to_csv('d11_cleaned_5min_2018_10_31.csv')
```
